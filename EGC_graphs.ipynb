```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
path = 'Графики ЭКГ'
file_list = os.listdir(path)
file_list
```

```python
a = pd.read_csv(os.path.join(path, file_list[0]))
dT=4*10**(-6)
b=list(range(len(a)))
t=map(lambda x: x*dT, b)
plt.figure(figsize=(10, 4), dpi=200)
plt.plot(list(t), a)
plt.title('Тестовый сигнал, I отв, 60 уд/мин, 2 мВ')  
plt.xlabel('t, сек')
plt.ylabel('U, мВ')
plt.grid()
```
